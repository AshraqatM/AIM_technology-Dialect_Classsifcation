# Data Cleaning and Preprocessing

Remove username

```python
data['text'] = data['text'].apply(lambda x: re.sub('@[^\s]+', '', x))
data
```

Remove Hashrags #

```python
data['text']= data['text'].apply(lambda x: re.sub('#', '', x) )
data  
```


Remove English characeters

```python
# data['text']= data['text'].apply(lambda x: re.sub('((www\.[^\s]+)|(https?://[^\s]+))','',x))
data['text']= data['text'].apply(lambda x: re.sub('[a-zA-Z\s]+',' ',x))
data  
```

Remove RT (Retweet)

```python
data['text']= data['text'].apply(lambda x: re.sub('RT','',x))
data 
```

Remove emojis

```python
emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+")   

data['text']= data['text'].apply(lambda x: emoji_pattern.sub(r'', x) )
data  
```

Remove digits

```python
from string import digits

data['text']= data['text'].apply(lambda x: x.translate(str.maketrans('', '', digits)) )
data 
```

Remove punctuations

```python
import string
# data['text']= data['text'].apply(lambda x: x.translate(str.maketrans('','',string.punctuation)) )
# data 

arabic_punctuations = '''`÷×؛<>_()*&^%][ـ،/:"؟.,'{}~¦+|\!”…“–ـ'''
english_punctuations = string.punctuation
punctuations_list = arabic_punctuations + english_punctuations

data['text']= data['text'].apply(lambda x: x.translate(str.maketrans('','',punctuations_list)) )
data 
```

Normalization

```python
data['text']= data['text'].apply(lambda x: re.sub("[إأآا]", "ا", x) )
data['text']= data['text'].apply(lambda x: re.sub("ى", "ي", x) )
data['text']= data['text'].apply(lambda x: re.sub("ة", "ه", x) )
data['text']= data['text'].apply(lambda x: re.sub("گ", "ك", x) )
data
```

Remove repeating characters

```python
data['text']= data['text'].apply(lambda x: re.sub(r'(.)\1+', r'\1', x) )
data
```

Remove single character

```python
data['text']= data['text'].apply(lambda x: re.sub(r'\b[A-Za-zÀ-ž]\b',' ', x) )
data
```

Remove stop words

```python
!pip install Arabic-Stopwords

import arabicstopwords.arabicstopwords as stp
stop_words = list(stp.classed_stopwords_list())
# #stp.is_stop(u'منكم')

# def remove_stop_words(corpus):
#     results = []
#     for text in corpus:
#         tmp = text.split(' ')
#         for stop_word in stop_words:
#             if stop_word in tmp:
#                 tmp.remove(stop_word)
#         results.append(" ".join(tmp))
        
#     return results

data['text']= data['text'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
data
```

Stemming

```python
from nltk.stem.isri import ISRIStemmer
st = ISRIStemmer()

# lis = ["يأكل","يفعل" ]
# lis2 = list()
# for i in lis:
#   lis2.append(st.stem(i))
#   print(lis2)

data['text']= data['text'].apply(lambda x: st.stem(x))
data
```

Tokenization

```python
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'\w+')
data['text'] = data['text'].apply(tokenizer.tokenize)
data
```
