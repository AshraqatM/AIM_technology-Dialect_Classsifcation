# Reading original dataset and adding the text to it then saving the new dataset

```python
df = pd.read_csv('dialect_dataset.csv')
df
```

POST request

```python
import requests

url = 'https://recruitment.aimtechnologies.co/ai-tasks'

text = list()
for i in range(0,len(df),1000):
  json_data = list(map(str,df[df.columns[0]].tolist()[i:i+1000]))
  r = requests.post(url, json=json_data)
  text.append(list(r.json().values()))
```

```python
#To make the list of equal sized arrays: we add none at the end (in order to reshape)
text[-1] = text[-1] + [None]*803
text_arr = np.array(text).reshape(459000)
text_arr = np.delete(text_arr, np.s_[-803:])
```

```python
#Adding fetched text to the dataset
df['text'] = text_arr

#drop id  column
df = df.drop('id', axis = 1)
df
```

Save Dataframe

```python
df.to_csv('new_dataset.csv')
df
```
